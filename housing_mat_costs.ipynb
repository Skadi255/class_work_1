```python
skip_rows = 2 
df = pd.read_csv('bldg_mat_change1.csv', skiprows=skip_rows)
df.head(31)
```

```python
df.columns
```

```python
# Convert the "Month" column to a datetime format
df['Month'] = pd.to_datetime(df['Month'], format='%b-%y')

# Convert the "All new construction" column to strings
df['All new construction'] = df['All new construction'].astype(str)

# Set the figure size
plt.figure(figsize=(12, 6))

# Plot the data
plt.plot(df['Month'], df['All new construction'], marker='o', color='b')

# Set the x and y-axis limits
plt.xlim(df['Month'].min(), df['Month'].max())  # Ensure the x-axis includes the entire time period
plt.ylim(0, 30)  # Set the y-axis limits to 0-30

# Set labels for the x and y axes
plt.xlabel('Month')
plt.ylabel('Percent Change')
plt.title('All New Construction Over Time')

# Rotate the x-axis labels for better visibility
plt.xticks(rotation=45)

# Show the plot
plt.grid(True)
plt.show()
```

```python
skip_rows = 1 
df = pd.read_csv('bldg_mat_change2.csv', skiprows=skip_rows)
df
```

```python
# Create a subset of the DataFrame with the relevant columns
df_subset = df[['Material or component', 'December 2019 to December 2020', 'January 2021 to July 2021']]

# Set the figure size
plt.figure(figsize=(12, 6))

# Extract data for plotting
x = df_subset['Material or component']
y1 = df_subset['December 2019 to December 2020']
y2 = df_subset['January 2021 to July 2021']

# Create an index for the x-axis positions
index = range(len(x))

# Plot the data as bars
plt.bar(index, y1, width=0.4, label='December 2019 to December 2020', align='center', alpha=0.7)
plt.bar([i + 0.4 for i in index], y2, width=0.4, label='January 2021 to July 2021', align='center', alpha=0.7)

# Set labels for the x and y axes
plt.xlabel('Material or Component')
plt.ylabel('Percentage Change')
plt.title('Percentage Change in Materials or Components')

# Set x-axis labels and rotate them for better visibility
plt.xticks([i + 0.2 for i in index], x, rotation=45)

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Create a subset of the DataFrame with the relevant columns
df_subset = df[['Material or component', 'December 2019 to December 2020', 'January 2021 to July 2021']]

# Set the figure size
plt.figure(figsize=(12, 6))

# Extract data for plotting
x = df_subset['Material or component']
y1 = df_subset['December 2019 to December 2020']
y2 = df_subset['January 2021 to July 2021']

# Create an index for the x-axis positions
index = range(len(x))

# Plot the data as bars
plt.bar(index, y1, width=0.4, label='December 2019 to December 2020', align='center', alpha=0.7)
plt.bar([i + 0.4 for i in index], y2, width=0.4, label='January 2021 to July 2021', align='center', alpha=0.7)

# Set labels for the x and y axes
plt.xlabel('Material or Component')
plt.ylabel('Percentage Change')
plt.title('Percentage Change in Materials or Components')

# Set x-axis labels and rotate them for better visibility
plt.xticks([i + 0.2 for i in index], x, rotation=45)

# Format the y-axis labels as percentages
def percentage_formatter(x, pos):
    return f'{x:.0%}'

y_format = FuncFormatter(percentage_formatter)
plt.gca().yaxis.set_major_formatter(y_format)

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
